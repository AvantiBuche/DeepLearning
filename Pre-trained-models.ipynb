```python
#ResNet-50 is 50 layers deep and is trained on a million images of 1000 categories from the ImageNet database.
#QFurthermore the model has over 23 million trainable parameters,
#which indicates a deep architecture that makes it better for image recognition

import numpy as np
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing.image import load_img,img_to_array
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
```

```python
model = ResNet50(weights='imagenet')
```

```python
model.summary()
```

```python
test_img = load_img('old-car.jpg',target_size=(224,224))
test_img
```

```python
x = img_to_array(test_img)
```

```python
x.shape
```

```python
x = x.reshape(1,224, 224, 3)
```

```python
x.shape
```

```python
preds = model.predict(x)
```

```python
preds.round(2)
```

```python
np.argmax(preds)
```

```python
print('Predicted:', decode_predictions(preds, top=3)[0])
```
