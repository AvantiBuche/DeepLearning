```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
import keras
from keras.layers import Dense
from keras.models import Sequential
dataset = pd.read_csv('iris.csv',names=['sep_l','sep_w','pet_l','pet_w','label'])
dataset
```

```python
X = dataset.iloc[:,0:4].values
X
```

```python
y = dataset.iloc[:,4].values
y
```

```python
#example 1 for LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() 
y = le.fit_transform(y)
#to map or transform data 
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
X_train.shape
```

```python
X_test.shape
```

```python
#one type of syntax
model = Sequential([
    Dense(10,activation="relu",input_shape=[4]),
    Dense(10,activation="relu"),
    Dense(3,activation="softmax")])
```

```python
model.summary()
```

```python
#parse cross because of label_endoder is used
model.compile(loss="sparse_categorical_crossentropy",optimizer='sgd',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
y_pred.round()
```

```python
#map data or parse in 1 digit
y_pred = np.argmax(y_pred,axis=1)
y_pred
```

```python
y_test
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
X1 = dataset.iloc[:,0:4].values
y1 = dataset.iloc[:,[4]].values
#store y data in array for OneHotEncoder
y1
```

```python
#example 2 for OneHotEncoder
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
y1 = ohe.fit_transform(y1)
y1
```

```python
X1_train,X1_test,y1_train,y1_test = train_test_split(X1,y1,test_size=0.2,random_state=0)
```

```python
#2nd type of syntax
model1 = Sequential()
model1.add(Dense(10,activation="relu",input_dim=4))
model1.add(Dense(10,activation="relu"))
model1.add(Dense(3,activation="softmax"))
```

```python
model1.summary()
```

```python
model1.compile(loss="categorical_crossentropy",optimizer='sgd',metrics=['accuracy'])
```

```python
model1.fit(X1_train,y1_train,epochs=100)
```

```python
model1.evaluate(X1_test,y1_test)
```

```python
y1_pred = model1.predict(X1_test)
y1_pred.round()
```

```python
y1_test = np.argmax(y1_test,axis=1)
y1_pred = np.argmax(y1_pred,axis=1)
```

```python
y1_test
```

```python
y1_pred
```

```python
print(confusion_matrix(y1_test,y1_pred))
```
